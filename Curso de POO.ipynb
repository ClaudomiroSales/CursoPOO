```python
#ifndef GRADEBOOK_H
#define GRADEBOOK_H

#include <iostream>
using std::cout;

#include <string>
using std::string;

class GradeBook
{
public:
    //GradeBook();
    //~GradeBook();

    void setCourseName( string name )
    {
        courseName = name;
    }

    string getCourseName(  )
    {
        return courseName;
    }

    void displayMessage( ) const
    {
        cout << "Welcome to the Grade Book for course " << courseName << '\n';
    }

private:

    string courseName;

};

#endif // GRADEBOOK_H
```

```python
#include<iostream>
using std::cout;
using std::cin;
using std::getline;

#include "GradeBook.h"

//This is where the execution of program begins
int main()
{
   
   GradeBook gradebook;
   string nameOfCourse;

   cout << "Digite o nome do curso: ";
   getline( cin, nameOfCourse );
   gradebook.setCourseName( nameOfCourse );


   gradebook.displayMessage( );

   
   return 0;
}
```

```python
#ifndef GRADEBOOK_H
#define GRADEBOOK_H

#include <iostream>
using std::cout;

#include <string>
using std::string;

class GradeBook
{
public:
    
    GradeBook( string name )
    {
        setCourseName( name );
    }
    //~GradeBook();

    void setCourseName( string name )
    {
        courseName = name;
    }

    string getCourseName(  )
    {
        return courseName;
    }

    void displayMessage( ) const
    {
        cout << "Welcome to the Grade Book for course " << courseName << '\n';
    }

private:

    string courseName;

};

#endif // GRADEBOOK_H
```

```python
#include<iostream>
using std::cout;
using std::cin;
using std::getline;

#include "GradeBook.h"

//This is where the execution of program begins
int main()
{
   
   GradeBook gradebook1( "Math" );
   GradeBook gradebook2( "Physics" );
   
   
   gradebook1.displayMessage( );
   gradebook2.displayMessage( );

   cout << "Os cursos sao: " << gradebook1.getCourseName( ) << " e " << gradebook2.getCourseName( ) << '\n';
   
   
   return 0;
}
```
